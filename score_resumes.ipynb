```python
import spacy
import pandas as pd
import json
from collections import defaultdict
```

```python
def check_testing_cat(label, text, score):
	if label == "Designation":
		if "test" in text:
			score['Testing'] += 1

	elif label == "Skills":
		if "test" in text:
			score['Testing'] += 1

	elif label == "Degree":
		if "test" in text:
			score['Testing'] += 1
```

```python
def check_development_cat(label, text, score):
	if label == "Designation":
		if "develop" in text:
			score['development'] += 1

	elif label == "Skills":
		if "develop" in text:
			score['development'] += 1

	elif label == "Degree":
		if "develop" in text:
			score['development'] += 1
```

```python
def check_management_cat(label, text, score):
	if label == "Designation":
		if "manage" in text:
			score['Management'] += 1

	elif label == "Skills":
		if "manage" in text:
			score['Management'] += 1

	elif label == "Degree":
		if "manage" in text:
			score['Management'] += 1
```

```python
nlp_model_path = "nlp_model/"
nlp_model = spacy.load(nlp_model_path) 
df = pd.DataFrame()

profiles = []

with open('clean_data/clean_data.json', 'r') as f:
	for f_line in f.readlines():
		resume = defaultdict(list , **json.loads(f_line))
		text_resume = resume['content'].replace("\n", " ").replace("\t", " ").strip()
		doc_r = nlp_model(text_resume)

		resume_score = {'Testing':0, 'development':0, 'Management':0}

		for ent in doc_r.ents:
			check_testing_cat(ent.label_, ent.text, resume_score)
			check_development_cat(ent.label_, ent.text, resume_score)
			check_management_cat(ent.label_, ent.text, resume_score)

		profiles.append(resume_score)

df_resumes_score = pd.DataFrame(profiles)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df_resumes_score)
```
